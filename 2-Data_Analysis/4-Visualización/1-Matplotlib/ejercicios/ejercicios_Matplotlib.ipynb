9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
import matplotlib.pyplot as plt

# Nombres de las personas
nombres = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']

# Alturas correspondientes en cent칤metros
alturas = [160, 180, 175, 190, 170]

# Crear el gr치fico de barras
plt.figure(figsize=(8, 6))
barplot = plt.bar(nombres, alturas, color='green')

# A침adir etiquetas encima de cada barra
for bar in barplot:
    altura = bar.get_height()  # Obtener la altura de la barra
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Posici칩n x centrada en la barra
        altura,  # Posici칩n y justo encima de la barra
        f'{altura} cm',  # Texto que muestra la altura
        ha='center',  # Alineaci칩n horizontal centrada
        va='bottom'   # Alineaci칩n vertical en la parte inferior del texto
    )

# Etiquetas y t칤tulo
plt.xlabel('Personas')
plt.ylabel('Altura (cm)')
plt.title('Altura de un Grupo de Amigos')

# Mostrar el gr치fico
plt.show()
```

17. Otro departamento B ha vendido en esos a침os 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gr치fica las dos l칤neas. B es una l칤nea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame para el departamento A
data_A = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000]
}
df_A = pd.DataFrame(data_A)

# Crear el DataFrame para el departamento B
data_B = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}
df_B = pd.DataFrame(data_B)

# Unir ambos DataFrames en uno solo
df = pd.merge(df_A, df_B, on='Year')

# Crear el gr치fico de l칤nea
plt.figure(figsize=(10, 6))  # Tama침o de la figura

# Graficar las l칤neas para cada departamento
plt.plot(df['Year'], df['Sold_items_A'], color='green', linestyle='--', marker='o', label='Departamento A')  # L칤nea para A
plt.plot(df['Year'], df['Sold_items_B'], color='red', linestyle=':', marker='x', label='Departamento B')  # L칤nea para B

# T칤tulo y etiquetas
plt.title('Ventas de Art칤culos por A침o')
plt.xlabel('A침o')
plt.ylabel('Art칤culos Vendidos')

# Mostrar la leyenda en la esquina inferior derecha
plt.legend(loc='lower right')

# Mostrar el gr치fico
plt.grid()  # A침adir una cuadr칤cula para mejor visualizaci칩n
plt.show()
```

20. Prueba a quitar la leyenda incluyendo legend igual a False, a침ade labels y quita la etiqueta en Y

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame para el departamento A
data_A = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000]
}
df_A = pd.DataFrame(data_A)

# Crear el gr치fico de pastel con solo los a침os como etiquetas
plt.figure(figsize=(8, 8))  # Tama침o de la figura
wedges, texts, autotexts = plt.pie(
    df_A['Sold_items_A'],
    labels=df_A['Year'],  # Usar solo los a침os como etiquetas
    startangle=140,
    autopct='',  # No mostrar los porcentajes
    textprops={'fontsize': 12},  # Tama침o de fuente para las etiquetas
    wedgeprops=dict(edgecolor='black')  # A침adir borde negro a las porciones
)

# Aumentar la legibilidad de las etiquetas
for text in texts:
    text.set_fontsize(12)  # Cambiar el tama침o de la fuente a 12
    text.set_color('black')  # Cambiar el color de la fuente a negro

# T칤tulo
plt.title('Distribuci칩n de Art칤culos Vendidos - Departamento A')

# Asegurarse de que el gr치fico sea un c칤rculo
plt.axis('equal')

# Cambiar el fondo del gr치fico a blanco
plt.gcf().patch.set_facecolor('white')

# Mostrar el gr치fico
plt.show()

```
