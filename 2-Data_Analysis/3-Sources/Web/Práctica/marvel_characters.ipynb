```python
def hash_params(timestamp, pub_key, priv_key):
    """ Marvel API requires server side API calls to include
    md5 hash of timestamp + public key + private key """

    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()

    return hashed_params

timestamp = datetime.datetime.now().strftime('%Y-%m-%d%H:%M:%S')

pub_key = '3c5df36b4dc254a1711ab795c56916b3'
priv_key = '0f4730af1920521244595e648e7261f63e6a8be8'

h_p = hash_params(timestamp, pub_key, priv_key)
h_p
```

```python
import requests
import hashlib
import time
from requests.adapters import HTTPAdapter
from urllib3.poolmanager import PoolManager
import ssl

# Custom Adapter para forzar el uso de TLS
class TLSAdapter(HTTPAdapter):
    def init_poolmanager(self, *args, **kwargs):
        context = ssl.create_default_context(ssl.Purpose.CLIENT_AUTH)
        context.options |= ssl.OP_NO_TLSv1 | ssl.OP_NO_TLSv1_1  # Deshabilita TLS 1.0 y 1.1
        kwargs['ssl_context'] = context
        return super(TLSAdapter, self).init_poolmanager(*args, **kwargs)

# Reemplaza estas variables con tus claves
public_key = "3c5df36b4dc254a1711ab795c56916b3"
private_key = "0f4730af1920521244595e648e7261f63e6a8be8"

# Genera el timestamp
timestamp = str(time.time())

# Crea el hash md5 con timestamp, private key y public key
hash_string = timestamp + private_key + public_key
hash_md5 = hashlib.md5(hash_string.encode('utf-8')).hexdigest()

# URL base de la API de Marvel
base_url = "http://gateway.marvel.com/v1/public/creators"

# Parámetros para la solicitud
params = {
    "nameStartsWith" : "L",
    "ts": timestamp,
    "apikey": public_key,
    "hash": hash_md5
}

# Configura la sesión con el Adaptador TLS
session = requests.Session()
session.mount('http://', TLSAdapter())

# Realiza la solicitud GET a la API de Marvel
response = session.get(base_url, params=params)

# Verifica el resultado
if response.status_code == 200:
    data = response.json()
    print("Datos obtenidos:")
    print(data)
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
marvel = data["data"]["results"]
marvel
id = []
fullName = []
picture_url = []

for personaje in marvel:
    id.append(personaje["id"])
    fullName.append(personaje["fullName"])
    picture_url.append(personaje["thumbnail"]["path"]+"."+personaje["thumbnail"]["extension"])

marvel_df = pd.DataFrame({"id": id, "fullName": fullName, "picture_url":picture_url })
marvel_df
```
