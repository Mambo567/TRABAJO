```python
# Conectamos con la base de datos chinook.db
connection = sqlite3.connect("chinook.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

### 1. Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la factura y el pais de la factura*

```python
query = '''
SELECT invoices.InvoiceId, customers.FirstName, customers.LastName,  invoices.InvoiceDate, Customers.Country

FROM invoices
INNER JOIN customers

ON invoices.CustomerId = customers.CustomerId

WHERE customers.Country = "Brazil"

ORDER BY invoices.InvoiceId


LIMIT 1000
'''

sql_query(query)
```

### 2. Facturas de Clientes de Brasil*

```python
query = '''
SELECT *

FROM invoices
INNER JOIN customers

ON invoices.CustomerId = customers.CustomerId

WHERE customers.Country = "Brazil"

ORDER BY InvoiceId ASC


LIMIT 1000
'''

sql_query(query)
```

### 3. Muestra cada factura asociada a cada agente de ventas con su nombre completo*

```python
query = '''
SELECT invoices.InvoiceId, employees.FirstName, employees.LastName, employees.EmployeeId

FROM invoices
INNER JOIN customers
ON invoices.CustomerId = customers.CustomerId
INNER JOIN employees
ON customers.supportRepid = employees.EmployeeId

ORDER BY invoices.InvoiceId ASC


LIMIT 1000
'''

sql_query(query)
```

### 4. Muestra el nombre del cliente, el pais, el nombre del agente y el total

```python
query = '''
SELECT customers.FirstName, customers.LastName, customers.Country, employees.FirstName, Employees.LastName, SUM(invoices.Total) AS TOTAL

FROM invoices
INNER JOIN customers
ON invoices.CustomerId = customers.CustomerId
INNER JOIN employees
ON customers.supportRepid = employees.EmployeeId
INNER JOIN invoice_items
ON invoices.InvoiceId = invoice_items.InvoiceId

GROUP BY customers.FirstName, customers.LastName



'''

sql_query(query)
```

### 5. Muestra cada articulo de la factura con el nombre de la cancion

```python
query = '''
SELECT Invoices.InvoiceId, tracks.Name

FROM invoices
INNER JOIN invoice_items
ON invoices.InvoiceId = invoice_items.InvoiceId
INNER JOIN tracks
ON invoice_items.TrackId = tracks.TrackId

ORDER BY Invoices.InvoiceId


'''

sql_query(query)
```

### 6. Muestra todas las conciones con su nombre, formato, album y genero

```python
query = '''
SELECT tracks.Name, media_types.Name, albums.Title, genres.Name

FROM tracks
INNER JOIN media_types
ON tracks.MediaTypeId = media_types.MediaTypeId
INNER JOIN albums
ON tracks.AlbumId = albums.AlbumId
INNER JOIN genres
ON tracks.GenreId = genres.GenreId



'''

sql_query(query)
```

### 7. Muestr a cuantas canciones hay en cada playlist y el nombre de cada playlist

```python
query = '''
SELECT playlists.Name, COUNT(tracks.Name)

FROM playlist_track
INNER JOIN tracks
ON playlist_track.TrackId = tracks.TrackId
INNER JOIN playlists
ON playlist_track.PlaylistId = playlists.PlaylistId

GROUP BY playlists.Name



'''

sql_query(query)
```

### 8. Muestra cuánto ha vendido cada empleado

```python
query = '''
SELECT employees.FirstName, employees.LastName, employees.Title, SUM(invoices.Total) AS TOTAL_SALES

FROM invoices
INNER JOIN customers
ON invoices.CustomerId = customers.CustomerId
INNER JOIN employees
ON customers.supportRepid = employees.EmployeeId

GROUP BY employees.FirstName, employees.LastName



'''

sql_query(query)
```

### 9. Quien ha sido el agente de ventas que más ha vendido en 2009?

```python
query = '''
SELECT employees.FirstName, employees.LastName, employees.Title, SUM(invoices.Total) AS CRACK

FROM invoices
INNER JOIN customers
ON invoices.CustomerId = customers.CustomerId
INNER JOIN employees
ON customers.supportRepid = employees.EmployeeId

WHERE invoices.InvoiceDate LIKE "2009%"

GROUP BY employees.FirstName, employees.LastName

ORDER BY SUM(invoices.Total) DESC

LIMIT 1


'''

sql_query(query)
```

### 10. Quien es son los 3 grupos que más han vendido?

```python
query = '''
SELECT artists.Name, SUM(invoices.Total)

FROM invoices
INNER JOIN invoice_items
ON invoices.InvoiceId = invoice_items.InvoiceId
INNER JOIN tracks
ON invoice_items.TrackId = tracks.TrackId
INNER JOIN albums
ON tracks.AlbumId = albums.AlbumId
INNER JOIN artists
ON albums.ArtistId = artists.ArtistId

GROUP BY artists.Name

ORDER BY SUM(invoices.Total) DESC

LIMIT 3


'''

sql_query(query)
```
