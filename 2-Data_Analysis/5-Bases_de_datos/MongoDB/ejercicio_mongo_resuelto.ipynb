```python
# 5. Devolver los restaurantes que se ubican en un valor de latitud inferior a -95.754168.

filtro = {"address.coord" : {"$lt": -95.754168}}
proyeccion = {"_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 6. Devolver los restaurantes que no preparan cocina americana y lograron una puntuación superior a 70 y se ubicaron en una longitud inferior a -65.754168.
filtro = {"cuisine" : {"$ne": "American "}, "grades.score" : {"$gt" : 70}, "address.coord.0" : {"$lt" : -65.754168}}
proyeccion = {"_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 7. Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificación 'A' que no pertenece al distrito de Brooklyn.  El documento debe mostrarse según la cocina en orden descendente.
filtro = {"cuisine" : {"$ne": "American "}, "grades.grade" : "A", "borough" : {"$ne" : "Brooklyn"}}
proyeccion = {"_id": 0}
list(restaurantes.find(filtro, proyeccion).limit(5).sort("cuisine",pymongo.DESCENDING))
```

```python
# 8. Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos.
filtro = {"borough" : "Bronx", "$or" : [{"cuisine" : "American "}, {"cuisine" : "Chinese"}]}
proyeccion = {"_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 9. Devolver `ID` del restaurante, `nombre`, `distrito` y la `cocina` para aquellos restaurantes que pertenecen al distrito de Staten Island, Queens, Bronx o Brooklyn.
filtro = {"$or" : [{"borough" : "Brooklyn"}, {"borough" : "Staten Island"}, {"borough" : "Queens"}, {"borough" : "Bronx"}]}
proyeccion = {"_id": 0, "restaurant_id" : True, "name" : True,"borough" : True, "cuisine" : True}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 10. Devolver `ID` del restaurante, `nombre`, `distrito` y la `cocina` de aquellos restaurantes que lograron una puntuación que no supere los 10.
filtro = {"grades.score" : {"$lt" : 10}}
proyeccion = {"_id": 0, "restaurant_id" : True, "name" : True,"borough" : True, "cuisine" : True}
list(restaurantes.find(filtro, proyeccion))
```

```python
from datetime import datetime, timezone
# 11. Devolver `ID` del restaurante, `nombre` y las `calificaciones` del restaurante para aquellos restaurantes que obtuvieron una calificación de "A" y un puntaje de 11 en una fecha ISO `"2014-08-11T00:00:00Z"` entre muchas fechas de encuesta.
fecha_iso = int(datetime(2014, 8, 11, 0, 0, 0, tzinfo=timezone.utc).timestamp() * 1000)
filtro = {"grades.grade" : "A", "grades.score" : 11, "grades.date.$date" : fecha_iso}
proyeccion = {"_id": 0, "restaurant_id" : True, "name" : True,"grades" : True}
list(restaurantes.find(filtro, proyeccion))
```

```python
# Alternativa
from datetime import datetime, timezone

# Convierte la fecha a timestamp en milisegundos
fecha_iso = int(datetime(2014, 8, 11, 0, 0, 0, tzinfo=timezone.utc).timestamp() * 1000)

# Filtro ajustado para buscar en grades.date.$date con el timestamp en milisegundos
filtro = {
    "grades": {
        "$elemMatch": {
            "grade": "A",
            "score": 11,
            "date.$date": fecha_iso  # Comparación con el timestamp en milisegundos
        }
    }
}

# Proyección para mostrar solo los campos necesarios
proyeccion = {
    "_id": 0,
    "restaurant_id": True,
    "name": True,
    "grades": True
}

# Ejecutar la consulta
resultados = list(restaurantes.find(filtro, proyeccion))
resultados
```
