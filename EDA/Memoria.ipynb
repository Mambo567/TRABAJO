# **Transformación Energética Global:** Identificando Oportunidades de Inversión en el Camino hacia la Sostenibilidad 
## **Tema**

Este proyecto ofrece un análisis exhaustivo de la transición energética global, centrándose en la comparación entre el crecimiento de las energías renovables y no renovables, y evaluando la eficiencia energética en sectores de alto consumo. A través de un enfoque integral que abarca la producción, el suministro y el consumo final de energía en economías clave, se identifican tendencias en la adopción de energías limpias y se destacan oportunidades estratégicas para mejorar la eficiencia energética. El análisis incluye proyecciones detalladas sobre los costes de generación para cada tipo de energía en 3 escenarios de politicas internacionales, con especial énfasis en los horizontes de 2030 y 2050, lo cual permite anticipar el impacto de las políticas y los avances tecnológicos en los precios de generación a largo plazo. Además, se examinan los flujos de inversión en infraestructura energética y las regiones y sectores con alto potencial para capitalizar en renovables y en la optimización de recursos energéticos. Este informe está diseñado para proporcionar a inversores y empresarios información estratégica que facilite la rentabilidad de la sostenibilidad y promueva decisiones informadas para maximizar el retorno de inversión en un contexto de transformación energética global.  
  
***Este proyecto tiene como objetivo responder a las preguntas clave que todo inversor en energía podría plantearse:***

- ***¿Cuál es la situación actual de la producción y el consumo de energía a nivel global?***
- ***¿Cuáles son los destinos más prometedores para invertir?***
- ***¿En qué sectores específicos resulta más estratégico orientar las inversiones?***
- ***¿Hacia dónde se están dirigiendo los flujos de inversión en la actualidad?***
- ***¿Qué tipos de energía presentan el mayor potencial de rentabilidad para invertir?***  



## **Hipótesis**  
1. La proporción de energías renovables en la producción y el consumo global de energía ha experimentado un incremento moderado en el año 2022 en comparación con el año 2000.
2. Europa y Estados Unidos destacan como los destinos más atractivos para la inversión en el sector energético.
3. Los sectores industrial y residencial representan los principales atractivos para la inversión en energías renovables, debido a su alto consumo y a las oportunidades significativas de mejora en eficiencia energética.
4. La mayor parte del capital de inversión en energías renovables se está dirigiendo a Europa.
5. Las energías renovables presentan ventajas considerables en costos de generación energética a largo plazo en comparación con las fuentes de energía tradicionales.  



## **Fuentes de datos**  
### ***International Energy Agency (IEA)*** / Link: https://www.iea.org/data-and-statistics/data-sets  

##### **Archivos utilizados:**
- **World Energy Balances Highlights 2024:** Este dataset de la IEA proporciona datos anuales sobre balances energéticos en países de todo el mundo, desglosando la producción, importación, exportación, suministro total y consumo final de energía por tipo (renovables y no renovables) desde 1971 hasta 2023 (provisional). Expresado en petajulios (PJ), permite analizar la evolución de la dependencia energética y la transición hacia energías renovables en cada región.  
- **IEA-EnergyEnd-usesandEfficiencyIndicatorsdatabase-HighlightsNovember2023:** Este dataset de la IEA proporciona datos sobre el consumo de energía en el sector residencial por país y tipo de fuente energética (como petróleo, gas, carbón y biocombustibles), desglosado por año de 2000 a 2021. Es útil para analizar la evolución en el uso de energías renovables y no renovables en hogares y su impacto en la eficiencia energética residencial. ​  
- **WorldEnergyInvestment2021_DataFile:** Este dataset de la IEA detalla la inversión global en energía desde 2015 hasta 2021, desglosada en categorías como combustibles fósiles, renovables, y otros tipos de suministro energético. Proporciona una visión general de cómo se distribuye el financiamiento en el sector energético, siendo útil para identificar tendencias de inversión en tecnologías limpias frente a fuentes convencionales.   
- **GECM_2024_Key input** Las siglas significan "Global Energy and Climate Model key input data". Se utiliza para analizar la transición energética global en términos de costos y proyecciones de generación. A través de sus datos clave, se pueden evaluar los costos de capital, costos operativos, factores de capacidad y el LCOE para distintas tecnologías de generación de energía en varios países y escenarios temporales.

Validar la **Hipótesis 2**: **Europa y Estados Unidos destacan como los destinos más atractivos para la inversión en el sector energético.**

Se observan los siguientes puntos clave:

- **China y Estados Unidos** son los mayores productores de energía a nivel mundial (De entre los paises incluidos en los datos).
- **Estados Unidos** destaca en la alta producción de **crudo y gas natural**.
- Se nota un crecimiento sustancial en el consumo energético de **China e India**, probablemente impulsado por su rápido crecimiento económico en las últimas dos décadas, mientras que **Estados Unidos y Europa** no han experimentado aumentos significativos en el consumo.

- **El gas natural** sigue siendo una fuente de energía significativa para **Estados Unidos y Europa**.

- Entre las principales fuentes de energía importadas por los cinco mayores importadores del mundo, el **petróleo crudo** ocupa el primer lugar.

Con base en lo anterior, se analiza:

Sin embargo, también importan grandes cantidades de energía, especialmente **crudo y carbón**, lo que indica que su producción interna no es suficiente para satisfacer su demanda. Esto los convierte en economías con una marcada **dependencia energética**. Además, ambos países dependen en gran medida de fuentes de energía de altas emisiones, como el carbón y el crudo. Lo anterior, añadido a su participación en acuerdos internacionales de reducción de emisiones ha incentivado a China e India a impulsar la **producción de energías renovables** para reducir su dependencia externa y sostener su crecimiento económico. Esta tendencia se refleja en el incremento significativo de la producción y consumo de energías renovables en ambos países. En conclusión, China e India representan economías emergentes con alto potencial de inversión en el sector energético, especialmente en energías renovables.

- **Europa**: Los datos señalan que Europa presenta una alta dependencia energética. Esto sumado a un fuerte compromiso con los acuerdos internacionales de reducción de emisiones, la convierte en una de las principales regiones interesadas en alcanzar la independencia energética mediante fuentes renovables. No obstante, Europa sigue dependiendo considerablemente de los derivados del petróleo en su consumo y exportaciones. En conjunto, Europa se percibe como una región con alto potencial para inversiones en energías renovables, impulsada por su considerable producción de energía limpia, su actual dependencia energética y su compromiso con la transición hacia fuentes sostenibles.

- **Estados Unidos**: Aunque comprometido con los acuerdos internacionales de reducción de emisiones, los datos señalan que Estados Unidos mantiene una economía fuertemente orientada hacia la producción, consumo y exportación de crudo, derivados del petróleo y gas natural. Esta estructura consolidada hace que la transición hacia energías renovables sea menos conveniente e inmediata que en otros países. Su **alta independencia energética** y el incremento notable en sus exportaciones de derivados del petróleo, crudo y gas natural entre 2000 y 2022 reflejan esta tendencia.

### **Propósito**
Validar la **Hipótesis 3**: **Los sectores industrial y residencial representan los principales atractivos para la inversión en energías renovables, debido a su alto consumo y a las oportunidades significativas de mejora en eficiencia energética.**

### **Análisis**
- En orden de consumo, los sectores con mayor demanda energética son **Industria** (especialmente en China e India), **Transporte** (principalmente en Estados Unidos y Europa) y **Residencial** (De manera general).
- **El carbón** sigue siendo la principal fuente de consumo energético en el sector industrial de **China** e **India**.
- La **energía renovable** representa la segunda mayor fuente de energía en el sector industrial de **India**. Sin embargo, el consumo de energía renovable en la industria de China es considerablemente bajo, lo cual puede significar que hay grandes obstáculos para una transición a energía limpia, lo cual también podria traducirse en oportunidades de inversión.
- El sector de **Transporte** en las cinco regiones depende sustancialmente de los **derivados del petróleo**.
- En el sector **Residencial**, el consumo de energía renovable es considerablemente alto en **China** e **India**.
- La **electricidad** es una fuente de energía sustancial en las cinco regiones, particularmente en los sectores **Industria, Residencial y de Servicios**; en estos dos últimos, es la fuente mayoritaria.
- A nivel global, entre todos los países incluidos en los reportes de la IEA, los sectores que más **emisiones** producen son, en orden: **Transporte, Industria, Residencial** y, por último, **Servicios**.
- A nivel global, entre todos los países incluidos en los reportes de la IEA, los sectores más **eficientes** son, en orden: **Industria, Transporte, Residencial** y, finalmente, **Servicios**.
- De entre todos los países incluidos en los reportes de la IEA, la región que genera más **emisiones** es **Estados Unidos**, seguida de **Europa**.
- De las tres regiones analizadas, el país con menor eficiencia energética es **Estados Unidos**, seguido de **Europa** y, por último, **Brasil**.  

**Global**: Los datos indican que los sectores de **Residencial** e **Industria** presentan el mayor potencial y apertura para la transición energética. En ambos sectores predomina el uso de **energía eléctrica** y existen avances en energías renovables, lo cual sugiere que la electricidad generada actualmente a partir de fuentes tradicionales podría reemplazarse por fuentes renovables. Por otro lado, el sector de **Transporte** depende en gran medida de los **derivados del petróleo** y es el principal productor de emisiones, lo que confirma que enfrenta los mayores retos en la transición energética. Esto podría representar oportunidades de inversión a largo plazo, especialmente cuando las presiones para reducir emisiones en el transporte se vuelvan más urgentes.

**China**: Los datos señalan que el sector **Industrial** en China utiliza muy poca **energía renovable** en contraste con el consumo significativo de energía limpia en el sector **Residencial**. Esto sugiere que el sector industrial aún depende del **carbón**. Sin embargo, dada su alta utilización de energía eléctrica, adoptar fuentes renovables para la generación de electricidad es viable y estratégico, considerando la presión interna y externa que China enfrentará para reducir sus emisiones industriales. Esto podría representar grandes oportunidades de inversión a mediano y largo plazo.

**India**: A diferencia de China, India muestra avances en el consumo de **energía renovable** tanto en el sector **Industrial** como en el **Residencial**, lo cual indica una apertura en el mercado energético y posiblemente una infraestructura renovable ya consolidada. Lo anterior recalca su potencial de inversión.

**Europa**: Europa presenta una alta dependencia del **gas natural**. Aunque el consumo de energías renovables ha crecido, aún es bajo en comparación con el consumo de electricidad generada a partir de fuentes tradicionales, lo cual destaca oportunidades de inversión para expandir la producción de energía renovable en la región.


### **Impacto en el Proyecto**

# **Sección 5: Proyecciones y análisis de costes de generación de energía**

### **Notas**  
Las visualizaciones se realizan bajo tres escenarios distintos, de los cuales el análisis se dirigirá principalmente a **STEPS** y **APS** por ser los más probables.

- **STEPS (Stated Policies Scenario)**: Proyecta el futuro en función de las políticas energéticas y climáticas vigentes, reflejando un avance moderado en la transición energética.
- **APS (Announced Pledges Scenario)**: Basado en los compromisos climáticos anunciados por los países, proyecta un escenario de transición más ambicioso, alineado con los objetivos de descarbonización.
- **NZE (Net Zero Emissions)**: Considera un escenario en el que se alcanzarían emisiones netas cero para mediados de siglo, impulsando una transición profunda y acelerada hacia energías limpias.

Estos escenarios permiten analizar distintos posibles caminos en función de políticas actuales, compromisos anunciados y objetivos de cero emisiones netas.


Este dataset en particular presenta un reto interesante y es que tiene mucha información valiosa en muchos escenarios, lo cual exige presentar la data de forma óptima.
### **Propósito**
Validar la **Hipótesis 5**: **Las energías renovables presentan ventajas considerables en costos de generación energética a largo plazo en comparación con las fuentes de energía tradicionales.**

### **Análisis**
- **STEPS**

  **Costes de capital**:  
  - La **energía solar** presenta los menores costes de capital a corto, mediano y largo plazo, lo que la hace muy atractiva para invertir tanto en la actualidad como en el futuro.
  - El **gas natural** se mantiene como la segunda fuente más competitiva para todas las regiones en todos los plazos.
  - Los costes de capital de la **energía eólica**, tanto en tierra como en mar, son competitivos. Sin embargo, en **India y China** podrían ser menos competitivos que el carbón y el gas natural tanto a corto como a largo plazo, mientras que en **Estados Unidos y Europa** se mantienen ligeramente por detrás del gas natural en términos de competitividad.
  - La **energía eólica marina** gana la mayor parte de su competitividad a largo plazo (2030 y 2050).

  **Costes Operativos**:  
  - Aunque el **gas natural** es competitivo en costes de capital, es considerablemente menos competitivo que las energías renovables en términos de costes operativos.
  - Las tres energías renovables destacan como las más competitivas en todos los plazos y para todas las regiones, especialmente en **Europa, India y China**.

  **Coste Total de Generación**:  
  - La **energía solar** presenta los costes más competitivos para todas las regiones en todos los plazos.
  - Le sigue la **energía eólica en tierra** como la segunda más competitiva.
  - La **energía eólica marina** aumenta su competitividad hacia el largo plazo (2050).
  - En general, las energías renovables son las más competitivas, con la solar y la eólica en tierra manteniendo este estatus desde el presente hasta el largo plazo. La eólica marina también se acerca a este nivel de competitividad en el largo plazo. La única fuente de energía tradicional que sigue siendo competitiva es el **carbón** (solo en India), aunque sigue siendo menos competitivo que las renovables.

- **APS**

  **Costes de capital**:  
  - En general, la **energía solar** se destaca como la más competitiva en todos los plazos y para todas las regiones.
  - El **gas natural** le sigue de cerca, manteniéndose competitivo en todos los plazos y para todas las regiones.
  - El **carbón** es poco competitivo en **Estados Unidos** y **Europa**, pero mantiene cierta competitividad en el corto y largo plazo en **China** e **India**, donde rivaliza con la competitividad de las energías eólicas en el corto plazo.
  - La **energía eólica marina** gana mayor competitividad en el largo plazo (2050), mientras que la eólica terrestre mantiene una competitividad media desde el corto plazo.

  **Costes Operativos**:  
  - Las tres energías renovables destacan como las más competitivas a nivel de todas las regiones y en todos los plazos.
  - El **carbón** y el **gas natural** son relativamente menos competitivos en el corto plazo y pierden todavía más competitividad en el largo plazo, quedando muy atrás de las energías renovables de forma general.

  **Coste Total de Generación**:  
  - En general, las tres energías renovables se destacan como las más competitivas a nivel de todas las regiones y en todos los plazos, especialmente la energía solar.
  - En **Estados Unidos** y **Europa**, el gas natural y el carbón son poco competitivos en el corto plazo. Además, en este escenario, se prevé su desuso para 2030 en el caso del carbón y para 2050 en el caso del gas natural.
  - En este escenario se plantea que **China** e **India** continúen utilizando carbón y gas natural hasta 2050. En estas regiones, el carbón es competitivo en el corto plazo, compitiendo con las energías eólicas. Sin embargo, en el largo plazo, ambas fuentes tradicionales pierden competitividad sustancialmente, volviéndolas inversiones poco atractivas.

**Conclusión**  
La **energía solar** se destaca como la fuente más competitiva en términos generales. Las **energías eólicas** son menos competitivas que el carbón y el gas en cuanto a inversión inicial (costes de capital), pero sus bajos costes de mantenimiento (costes operativos) y las previsiones de regulaciones cada vez más estrictas sobre emisiones (reflejadas en los costes totales de generación) las hacen sustancialmente más competitivas que el carbón y el gas a largo plazo.  
Por lo tanto, las energías más atractivas para invertir son la **solar** y la **eólica terrestre** en el corto plazo, y la **eólica marina** en el mediano y largo plazo. Las tres energías renovables presentan proyecciones de alta competitividad a largo plazo, con una ventaja que se fortalece con el tiempo.  
El **carbón** y el **gas natural** pierden competitividad de manera sustancial en el mediano y largo plazo, y en Estados Unidos y Europa ya está planeado su desuso.


### **Impacto en el Proyecto**
Se valida la **Hipótesis 5**, por lo que se concluye que **las energías renovables presentan ventajas considerables en costos de generación energética a largo plazo en comparación con las fuentes de energía tradicionales.** Esto da todavía mas fuerza al argumento del interés que tendrán las regiones en alcanzar la independencia energética a través de fuentes de energía renovables, especialmente **China, India y Europa.**



# **Conclusiones**

Con base en los análisis de las secciones de este EDA y una posterior investigación, se proponen las siguientes oportunidades de inversión:

#### 1. China: Oportunidad para Inversores Grandes y Experimentados
- **Perfil de Inversor**: Grandes corporaciones, fondos institucionales o inversores experimentados en mercados emergentes, con alta capacidad financiera y tolerancia a riesgos regulatorios y geopolíticos.
- **Ventajas**:
  - **Crecimiento continuo en renovables y mercado de gran escala**: China ofrece oportunidades únicas para proyectos de gran envergadura, en sectores como manufactura y tecnología renovable.
  - **Apoyo gubernamental a proyectos conjuntos (joint ventures)**: Invertir en asociación con empresas locales permite acceder a incentivos y reducir barreras regulatorias.
  - **Competitividad de Energía Solar y Eólica**: Las condiciones geográficas y climáticas en varias zonas del país favorecen la energía solar y eólica terrestre en el corto plazo. La energía eólica marina es prometedora para el mediano y largo plazo, especialmente en zonas costeras.
- **Consideraciones**:
  - **Regulación estricta y competencia local**: La regulación y las políticas de acceso al mercado pueden ser complejas, lo que hace fundamental contar con socios locales de confianza. La competencia con empresas nacionales también puede ser intensa.
- **Recomendación de Inversión**: Energía solar y eólica terrestre a corto plazo, y energía eólica marina a mediano y largo plazo, con un enfoque en los sectores **industrial** y **residencial**, que concentran una alta demanda energética.

#### 2. India: Atractivo para Inversores Medianos con Aversion al Riesgo Moderada
- **Perfil de Inversor**: Empresas medianas, fondos en crecimiento o inversores con una aversión moderada al riesgo, interesados en mercados de alto crecimiento y menor complejidad regulatoria que China.
- **Ventajas**:
  - **Demanda energética en expansión y rápido desarrollo de renovables**: India tiene un mercado interno creciente impulsado por la demanda de electricidad y políticas de sostenibilidad, lo que la convierte en un destino atractivo para proyectos de energía solar y eólica.
  - **Incentivos y apertura a inversión extranjera**: Con menos restricciones que China, India promueve la inversión en renovables, particularmente en energía solar y eólica terrestre.
  - **Condiciones geográficas favorables para la energía solar**: El alto nivel de radiación solar en gran parte del territorio indio favorece proyectos solares rentables, mientras que la energía eólica terrestre tiene buen potencial en ciertas regiones.
- **Consideraciones**:
  - **Desafíos de infraestructura y burocracia**: La infraestructura energética de India está en desarrollo, y aún enfrenta obstáculos burocráticos que pueden afectar los tiempos de retorno de la inversión.
- **Recomendación de Inversión**: Invertir en **energía solar y eólica terrestre** a corto plazo, con una visión hacia la eólica marina en el mediano-largo plazo. Los sectores **industrial** y **residencial** son particularmente atractivos por su demanda energética y oportunidades de crecimiento.

#### 3. Unión Europea: Opción Segura para Inversores Conservadores en un Mercado Competitivo
- **Perfil de Inversor**: Inversores conservadores, fondos de pensiones y otros inversores institucionales que priorizan la estabilidad en un entorno regulatorio seguro y competitivo.
- **Ventajas**:
  - **Estabilidad y regulaciones favorables a largo plazo**: La UE es un líder mundial en la transición energética, con políticas firmes de reducción de emisiones que crean un entorno seguro para la inversión en energías renovables.
  - **Fomento de tecnologías avanzadas y apoyo gubernamental**: La UE apoya proyectos de innovación en hidrógeno verde, almacenamiento de energía y otras tecnologías emergentes, lo cual representa oportunidades para inversiones de bajo riesgo con potencial de crecimiento.
  - **Geografía diversa favorable tanto para solar como para eólica**: La energía solar es especialmente viable en países como **España** e **Italia**, mientras que la energía eólica tiene alta competitividad en **Alemania** y **Reino Unido**; **Francia** representa un buen equilibrio para ambas fuentes.
- **Consideraciones**:
  - **Alta competencia y márgenes más ajustados**: La consolidación y madurez del mercado en Europa generan una competencia elevada y menores márgenes de retorno en comparación con mercados emergentes.
- **Recomendación de Inversión**: Invertir en **energía solar** para países mediterráneos como España e Italia y **energía eólica** en Alemania y Reino Unido, con una combinación de ambas en Francia. Foco en el sector **residencial** a corto plazo, seguido del **sector industrial** y, en el mediano/largo plazo, el **sector de transporte** que lidera en emisiones y probablemente, sea Europa el que tome inciaitivas con regulaciones en transición hacia energías limpias.

### Resumen
- **China**: Ideal para grandes inversores experimentados que puedan asumir riesgos en un mercado de alta escala y crecimiento, recomendándose energía solar y eólica terrestre a corto plazo y eólica marina a largo plazo.
- **India**: Atractiva para inversores medianos que busquen alto crecimiento en un mercado en desarrollo con menor complejidad regulatoria que China, siendo la energía solar y eólica terrestre las opciones más viables.
- **Unión Europea**: La mejor opción para inversores conservadores que valoren la seguridad en un mercado consolidado, con energía solar y eólica dependiendo del país. Europa ofrece estabilidad y apoyo a la innovación en tecnologías limpias, con oportunidades a corto plazo en los sectores residencial e industrial.

Este enfoque permite que cada perfil de inversor optimice sus oportunidades de inversión y minimice riesgos en función de sus necesidades y características, logrando una alineación estratégica con el potencial de crecimiento de cada región.
